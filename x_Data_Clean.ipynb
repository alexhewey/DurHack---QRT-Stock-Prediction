```python
import pandas as pd
import numpy as np
```

```python
test = pd.read_csv('x_train.csv')
print(test['VOLUME_1'].isnull().sum())
R_In = list(range(17,24,2))
V_In = list(range(18,25,2))
test = test.dropna(subset = test.columns[R_In], how = 'all')
test = test.dropna(subset = test.columns[V_In], how = 'all')
print(test['VOLUME_1'].isnull().sum())
```

```python
drp_ind=list(range(28,test.shape[1]))
test = test.drop(test.columns[drp_ind],axis = 1)
```

```python
for i in range(1,10):
    test['RET_%d_diff' %i] = test['RET_%d' %i]-test['RET_%d' %(i+1)]
    test['VOLUME_%d_diff' %i] = test['VOLUME_%d' %i]-test['VOLUME_%d' %(i+1)]
```

```python
new_features = []
shifts = range(1,10)  # Choose some different shifts\n",
statistics = ['mean']  # the type of stat\n",
gb_features = ['SECTOR', 'DATE']
target_feature_1 = 'RET'
target_feature_2 = 'diff'
tmp_name = '_'.join(gb_features)
for shift in shifts:
    for stat in statistics:
        name = f'{target_feature_1}_{shift}_{target_feature_2}_{tmp_name}_{stat}'
        feat = f'{target_feature_1}_{shift}_{target_feature_2}'
        new_features.append(name)
        for data in [test]:
            data[name] = data.groupby(gb_features)[feat].transform(stat)
            
```
